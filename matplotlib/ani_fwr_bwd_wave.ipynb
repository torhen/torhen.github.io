```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np


x = np.linspace(0, 2 * np.pi, 100)

fig = plt.figure(facecolor='white')
ax = fig.add_subplot()



def animate(frame):
    ax.cla()
    ax.set_title(f"frame {frame}")
    ax.set_ylim(-3, 3)
    ax.grid()
    y1 = np.sin(x - frame/5)
    y2 = np.sin(x + frame/5)
    ax.plot(x, y1, label='forward')
    ax.plot(x, y2, label='backward')
    ax.plot(x, y1 + y2, label='sum')
    ax.legend(loc="lower left")

plt.close(fig)
ani = FuncAnimation(fig, animate)  
# ani.save('wave.gif')

import IPython
IPython.display.HTML(ani.to_jshtml())

```
